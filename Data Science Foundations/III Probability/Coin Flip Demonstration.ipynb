# Coin Flip Demonstration



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def flip_coin():
    # create our coins as lists of outcomes
    coin_A = ['H','T']
    coin_B = ['H','T']

    # flip the coins randomly
    coin_A_result = np.random.choice(coin_A)
    coin_B_result = np.random.choice(coin_B)

    # we check if both flips are heads
    return coin_A_result == 'H' and coin_B_result == 'H'
```

```python
# tracking the trials
num_trials = 5
proportions = []
flips = []
```

```python
# track the number of times heads occurs twice
two_heads = 0
```

```python
# lets run the experiment five times'
for flip in range(num_trials):
    two_heads += flip_coin()

    # track the proportion 
    proportions.append(two_heads / (flip + 1))

    # track the number of flips
    flips.append(flip + 1)
```

```python
# plot everything
plt.plot(flips, proportions, label='Experimental Probability')
plt.xlabel('Number of Flips')
plt.ylabel('Proportion of Two Heads')
plt.hlines(0.25, 0, num_trials, colors='green', label='True Probability')
plt.legend()
plt.show()
plt.close()
```
